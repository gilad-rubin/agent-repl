```python
from IPython.display import display, HTML
import time

progress = display(HTML("<div style=\"width:300px;background:#eee;border-radius:4px\"><div id=\"bar\" style=\"width:0%;height:20px;background:#3b82f6;border-radius:4px;transition:width 0.3s\"></div></div><span id=\"pct\">0%</span>"), display_id="progress")

for i in range(1, 11):
    time.sleep(0.5)
    pct = i * 10
    progress.update(HTML(f"<div style=\"width:300px;background:#eee;border-radius:4px\"><div style=\"width:{pct}%;height:20px;background:#3b82f6;border-radius:4px;transition:width 0.3s\"></div></div><span>{pct}%</span>"))

print("Done!")
```
